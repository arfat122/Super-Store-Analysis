# Customer Acquisition

```python
Orders.groupby("Order ID").min('Order Date')
```

```python
Customer_Acquisition_by_date = Orders.groupby('Customer ID')['Order Date'].min().reset_index()
Customer_Acquisition_by_date['Year'] = Customer_Acquisition_by_date['Order Date'].dt.strftime("%Y")
Customer_Acquisition_by_date['Month'] = Customer_Acquisition_by_date['Order Date'].dt.strftime("%B")
Customer_Acquisition_by_date['Month Number'] = Customer_Acquisition_by_date['Order Date'].dt.month
```

```python
Customer_Acquisition_by_Year = Customer_Acquisition_by_date.groupby('Year')['Customer ID'].count().reset_index()
```

```python
sns.lineplot(x='Year',y='Customer ID',data = Customer_Acquisition_by_Year,marker = 'o')
plt.title('Customers Acquition By Year',color = 'g')
plt.xlabel('Year',fontsize = 12)
plt.ylabel('Customers',fontsize = 12)
```

```python
Customer_Acquisition_by_Year['CumSum'] = Customer_Acquisition_by_Year['Customer ID'].cumsum()
```

```python
sns.lineplot(x='Year',y='CumSum',data = Customer_Acquisition_by_Year,marker = 'o')
plt.title('Total Customer Acquition Till Date',color = 'g')
plt.xlabel('Year',fontsize = 12)
plt.ylabel('Customers',fontsize = 12)
```

```python
Customer_Acquisition_By_Year_month = Customer_Acquisition_by_date.groupby(['Year','Month','Month Number'])['Customer ID'].count().reset_index().sort_values(by='Month Number')
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x='Month',y = 'Customer ID' , data = Customer_Acquisition_By_Year_month ,hue = 'Year',palette = 'rocket')
```
